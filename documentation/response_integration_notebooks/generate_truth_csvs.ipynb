```python
meta_vars = [f'v_{x}' for x in 
             [5,6,11,118,119,120,121,12,14,15,16,19,124]]
rating_vars = [f'v_{x}' for x in range(7039,7474)]
overflow_vars = ['v_18','v_1373']
```

```python
meta_df = df[['lfdn']+meta_vars]
meta_df.head()
```

```python
def integrate_corrections(df_old, df_others, var1, var2):
    df_old = df_old.set_index('lfdn')
    df_old[f'v_{var1}_{var2}_integrated'] = df_old[f'v_{var1}']
    for row in df_others.iterrows():
        df_old.at[row[1]['lfdn'],f'v_{var1}_{var2}_integrated'] = row[1][f'v_{var2}_coded']
    return df_old.reset_index()

def add_coded_column(df_old, colvalues, var):
    df_old[f'v_{var}_coded'] = colvalues
    return df_old
```

Metadata Frame

```python
v_6_some_coded = pd.read_csv('../../data/codedtruth/v_6_coded_final.csv')
v_16_some_coded = pd.read_csv('../../data/codedtruth/v_16_coded_final.csv')
v_11_all_coded_cols = pd.read_csv('../../data/codedtruth/v_11_coded_final.csv')['v_11_coded']
v_19_all_coded_cols = pd.read_csv('../../data/codedtruth/v_19_coded_final.csv')['v_19_coded']
```

```python
meta_df_integrated = add_coded_column(
    add_coded_column(
        integrate_corrections(
            integrate_corrections(
                meta_df, v_6_some_coded, 5, 6), 
            v_16_some_coded, 15, 16), 
        v_11_all_coded_cols, 11), 
    v_19_all_coded_cols, 19)
```

```python
# this is necessary because I filtered out the 'Other (please specify)' answers that said 'NotAnswered' 
# before giving the responses to validators
meta_df_integrated['v_5_6_integrated'
                  ] = meta_df_integrated.v_5_6_integrated.replace('Other (please specify)', 'Other')
```
